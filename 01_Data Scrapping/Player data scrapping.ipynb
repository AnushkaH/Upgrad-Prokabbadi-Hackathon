```python
total_points = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_102_stats.json"
successful_raids = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_21_stats.json"
raid_points = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_22_stats.json"
successful_tackles = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_23_stats.json"
tackle_points = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_103_stats.json"
avg_raid_point = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_138_stats.json"
avg_tackle_points = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_139_stats.json"
dod_raid_points = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_132_stats.json"
super_raids = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_104_stats.json"
super_tackles = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_28_stats.json"
super_10s = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_100_stats.json"
high_5s = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_101_stats.json"

# sample = json_normalize(requests.get(high_5s).json()['data'])
# sample
```

```python
players_data = []
# for url in players_meta:
dataset = requests.get(total_points).json()['data']
for i in dataset:
    player_obj = {
                    'player_id': i.get('player_id', None),
                    'player_name': i.get('player_name', None),
                    'matches_played': i.get('match_played', None),
                    'team_name': i.get('team_full_name', None),
                    'total_points': i.get('value', None),
                    'successful_raids': None,
                    'raid_points': None,
                    'successful_tackles': None,
                    'tackle_points': None,
                    'avg_raid_point': None,
                    'avg_tackle_points': None,
                    'dod_raid_points': None,
                    'super_raids': None,
                    'super_tackles': None,
                    'super_10s': None,
                    'high_5s': None 
                }
    players_data.append(player_obj)
    
dataset = requests.get(successful_raids).json()['data']

points_meta = ['successful_raids', 
          'raid_points', 
          'successful_tackles', 
          'tackle_points', 
          'avg_raid_point', 
          'avg_tackle_points', 
          'dod_raid_points', 
          'super_raids',
          'super_tackles',
          'super_10s',
          'high_5s']
for l in points_meta:
    dataset = requests.get(locals()[l]).json()['data']
    for i in dataset:
        for j in players_data:
            if i['player_id'] == j['player_id']:
                j[l] = i['value']
            else:
                continue
        
total_player_points = json_normalize(players_data)

total_player_points.to_csv('player_data.csv', header= True)
```
